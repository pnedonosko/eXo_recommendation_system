# **Data preprocessing**

## **Drop redundant columns**

Here I'll drop redundant columns. Columns can be redundant i different ways, for instance they can be useless for model training (i.e. IDs), or thay can have zero correlation with target column and with whole dataset at all.



\*As target feature is synthetic, it is possible that some features, that should make a greate contribution to ranking of a posts (i.e. features related to the age of post) can have near to zero correlation, so here I'll just drop redundant features in terms of useless for model.

```python
df.shape
```

```python
# drop ids columns
df.drop(['id', 'owner_id', 'poster_id', 'participant1_id',
         'participant2_id', 'participant3_id',
         'participant4_id', 'participant5_id'], axis=1, inplace=True)
```

```python
# drop columns related to time, because some of them are useless or gathered badly
df.drop(['updated_time', 'age', 'updated_age'], axis=1, inplace=True)
```

```python
df.shape
```

So, I've dropped 11 columns, which don`t make a reasonable contribution to our model. This will speed up a training time and possibly can increase accuracy.

## **Normalizing (Scaling)**

Below you can see all numerical features that we have at this point:

```python
df.describe()
```

As we can see, some features are already scaled to the interval of [0,1], some are binary variables (0/1) and others are features denoting number of smth (number_of_likes/comments, etc.). Binary variables I'll convert to the categorical feature columns; features denoting  number of smth I'll bucketize in furter sections.

Summing up, we don`t really need normalizing here, all features values distributions are fine.

```python
#plot correlation matrix
plt.figure(figsize = (36,22))
plt.title('Correlation Heatmap', fontsize = 16);

# Mask to remove redundancy from the heatmap.
mask = np.zeros_like(df.corr(), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(df.corr(), mask=mask, vmin=-1, vmax = 1, cmap = "coolwarm",  annot = True);
```

```python
#plot correlation matrix
plt.figure(figsize = (36,22))
plt.title('Correlation Heatmap', fontsize = 16);

# Mask to remove redundancy from the heatmap.
mask = np.zeros_like(df.corr(method="spearman"), dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(df.corr(method="spearman"), mask=mask, vmin=-1, vmax = 1, cmap = "coolwarm",  annot = True);
```
